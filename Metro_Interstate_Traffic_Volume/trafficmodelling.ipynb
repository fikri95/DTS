```python
data1.drop(['tanggal','weather_main'],axis=1,inplace=True)
data1.head()
```

```python
data1['holiday'] = LabelEncoder().fit_transform(data1['holiday'])
data1['weather_description'] = LabelEncoder().fit_transform(data1['weather_description'])
data1['waktu'] = LabelEncoder().fit_transform(data1['waktu'])
```

```python
data_proses=data1
X=data_proses.drop(['traffic_volume'],axis=1)
Y=data_proses['traffic_volume']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.30)
```

```python
reg = {
    'Linear Regression' : LinearRegression(),
    'Decision Tree' : DecisionTreeRegressor(),
    'Random Forest' : RandomForestRegressor(),
    'Support Vector Machines' : SVR(),
    'K-nearest Neighbors' : KNeighborsRegressor(),
}
error=pd.DataFrame(columns=['MSE','R2-score','RMSE'])
for i,j in reg.items():
    j.fit(X_train,Y_train)
    pred = j.predict(X_test)
    error.loc[i]= [mean_squared_error(Y_test,pred),r2_score(Y_test,pred),mean_squared_error(Y_test,pred,squared=False)]
error
```

```python
error1=pd.DataFrame(columns=['MSE','R2-score','RMSE'])
for metod,fung in reg.items():
    Input = [('scale', StandardScaler()), ('model', fung)]
    pipe = Pipeline(Input)
    pipe.fit(X_train,Y_train)
    pred=pipe.predict(X_test)
    error1.loc[metod]= [mean_squared_error(Y_test,pred),r2_score(Y_test,pred),mean_squared_error(Y_test,pred,squared=False)]
error1
```
