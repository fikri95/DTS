```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
data=pd.read_csv('menu.csv')
data.head()
```

```python
x=data[['Item','Total Fat (% Daily Value)','Saturated Fat (% Daily Value)','Cholesterol (% Daily Value)', 
       'Sodium (% Daily Value)', 'Carbohydrates (% Daily Value)', 
       'Dietary Fiber (% Daily Value)','Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
       'Calcium (% Daily Value)', 'Iron (% Daily Value)']]
x.head()
```

```python
daily1=x.set_index('Item')
daily1.head()
```

```python
col=daily1.columns
col
```

```python
kombr={}
for j in col:
    komb=0
    for i in range(1,5):
        if komb!=0:
            break
        for el in itertools.combinations(daily1[j], i):
            if sum(el)==100:
                komb=i
                
    kombr[j]=komb
kombr
```

```python
df=pd.DataFrame(kombr.items(),columns=["Nutrition", "Minimum Number"])
df
```

```python
res=[]
ls = list(itertools.combinations(daily1.index, 3))
list1=[list(g) for g in ls]
ls1=[inx for inx in list1 if daily1['Iron (% Daily Value)'].loc[inx].sum() == 100] 
res.append(ls1)
res
```

```python
res1=[]
ls1 = list(itertools.combinations(daily1.index, 2))
list2=[list(g) for g in ls1]
ls2=[inx for inx in list2 if daily1['Total Fat (% Daily Value)'].loc[inx].sum() == 100] 
res1.append(ls2)
res1
```

```python
totalfat=pd.DataFrame(columns=['Total Fat (% Daily Value)', 'Saturated Fat (% Daily Value)',
       'Cholesterol (% Daily Value)', 'Sodium (% Daily Value)',
       'Carbohydrates (% Daily Value)', 'Dietary Fiber (% Daily Value)',
       'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
       'Calcium (% Daily Value)', 'Iron (% Daily Value)'])
totalfat.index.name='Kombinasi'
totalfat
```

```python
for i in res1[0]:
    idx=' '.join([str(x) for x in i])
    totalfat.loc[idx]=daily1.loc[i].sum(axis=0)
totalfat
```
